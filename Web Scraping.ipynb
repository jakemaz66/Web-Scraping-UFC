```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from webdriver_manager.microsoft import EdgeChromiumDriverManager
import time

driver = webdriver.Edge(service=Service(EdgeChromiumDriverManager().install()))

url = 'https://www.itnwwe.com/mma/ufc-fighters-roster/'
driver.get(url)

time.sleep(5)
tr_elements = driver.find_elements("xpath", "//tbody/tr")

elements_dict = {}

for tr_element in tr_elements:
    td_elements = tr_element.find_elements("xpath", ".//td")
    
    key = td_elements[0].text
    
    values = [td.text for td in td_elements[1:-1]]
    
    last_value = td_elements[-1].text.split('-')
    
    elements_dict[key] = values + last_value

print(elements_dict)
driver.quit()
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(elements_dict, orient='index')

csv_file = 'output.csv'
df.to_csv(csv_file, index_label='Key')

print(f"CSV file '{csv_file}' has been created.")
```
