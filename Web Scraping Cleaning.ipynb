```python
dataset = pd.read_csv(r"C:\Users\jakem\OneDrive\Desktop\Data Science Projects\07-2023 Web Scraping UFC\output.csv")
dataset.head()
```

```python
dataset = dataset.dropna(how='all')
```

```python
kamaru = dataset.iloc[36, 4]
kamaru
```

```python
dataset = dataset.replace('3Justin Gaethje', '3')
```

```python
dataset['1'].fillna('not specified', inplace=True)
dataset['0'].fillna('no nickname', inplace=True)
dataset['1'] = dataset['1'].replace('–', 'not specified')
dataset['0'] = dataset['0'].replace('–', 'no nickname')

dataset.head(55)
```
